```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
from pdb import set_trace
```

```python
data_path = '/Users/omarschall/personal/frisbee/STATS/passing_datums/'
```

```python
for i_file, file in enumerate(os.listdir(data_path)):

    if '.csv' not in file:
        continue
        
    if i_file == 0:
        
        df = pd.read_csv(os.path.join(data_path, file))
        
    else:
        
        df_ = pd.read_csv(os.path.join(data_path, file))
        df = df.append(df_)
```

```python
df = df.dropna(subset={'Thrower', 'Receiver'})
```

```python
players = list(set(df['Thrower']).union(set(df['Receiver'])))
```

```python
players
```

```python
players = list(sorted(players))
player_inits = []
for p in players:
    try:
        player_init = p.split(' ')[0][0] + p.split(' ')[1][0]
    except IndexError:
        player_init = p.split(' ')[0][:2]
    player_inits.append(player_init)
```

```python
player_inits
```

```python
df = df[df['Turnover?'] == 0]
df = df.reset_index()
```

```python
import matplotlib
n = len(players)
figs_path = '/Users/omarschall/personal/frisbee/STATS/figs'
fig = plt.figure(figsize=(18, 18))
matrix = np.zeros((n, n))
N_throws = np.zeros((n, n))
for i_pass in range(len(df)):
    
    thrower = df['Thrower'][df.index[i_pass]]
    receiver = df['Receiver'][df.index[i_pass]]
    #set_trace()
    i_player = players.index(thrower)
    j_player = players.index(receiver)
    
    N_throws[i_player, j_player] += 1
    matrix[i_player, j_player] += np.float32(df['Forward distance (m)'][df.index[i_pass]])
    
matrix = matrix/N_throws
matrix[np.isinf(matrix)] = 0
matplotlib.rc('image', cmap='RdBu_r')
current_cmap = matplotlib.cm.get_cmap()
current_cmap.set_bad(color='C1')

plt.imshow(matrix)#, cmap='RdBu_r')
plt.xticks(range(len(players)), player_inits, fontsize=18)
plt.yticks(range(len(players)), player_inits, fontsize=18)
plt.colorbar()
fig.savefig(os.path.join(figs_path, 'All_Games_Yards_Matrix.pdf'), dpi=300, format='pdf')
```
